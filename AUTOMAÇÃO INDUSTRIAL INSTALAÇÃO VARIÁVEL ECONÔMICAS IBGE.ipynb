```python
import plotly.graph_objects as go
import numpy as np
from scipy.stats import norm

# Definindo um valor específico para calcular a probabilidade
valor_especifico = 2500000  # Substitua pelo valor desejado

# Gerando dados para a distribuição normal
x = np.linspace(media_ind - 3 * desviopadrão_ind, media_ind + 3 * desviopadrão_ind, 1000)
y = norm.cdf(x, media_ind, desviopadrão_ind)

# Calculando a probabilidade para o valor específico
probabilidade = norm.cdf(valor_especifico, media_ind, desviopadrão_ind)

# Criando o gráfico de densidade de probabilidade acumulada (CDF)
fig_cen2500 = go.Figure()

# Adicionando a curva CDF
fig_cen2500.add_trace(go.Scatter(x=x, y=y, mode='lines', name='CDF'))

# Destacando o ponto específico no gráfico
fig_cen2500.add_trace(go.Scatter(x=[valor_especifico, valor_especifico], y=[0, probabilidade],
                         mode='lines', name=f'Probabilidade: {probabilidade:.4f}',
                         line=dict(color='red', dash='dash')))

# Layout do gráfico
fig_cen2500.update_layout(title=f'Probabilidade Cumulativa (CDF) - Valor Específico: {valor_especifico}',
                  xaxis_title='PIB da Indústria (Milhões de Reais)',
                  yaxis_title='Probabilidade Cumulativa',
                  font=dict(family="Arial", size=12, color="darkblue"),  # Font style for title and axis labels
                  title_font=dict(family="Arial", size=12, color="black"),  # Font style for title
                  paper_bgcolor="lightgray",  # Background color of the plot
                  plot_bgcolor="white",  # Background color of the plot area
                  showlegend=True)

# Exibição do gráfico
fig_cen2500.show()
```

# GERANDO A REGRESSÃO LINEAR MÍNIMOS QUADRADOS (ORDINARY LEAST SQUARE OLS)

A regressão por mínimos quadrados ordinários (OLS) é uma estratégia de otimização que ajuda a encontrar uma linha reta o mais próximo possível dos pontos de dados em um modelo de regressão linear. A OLS é considerada a estratégia de otimização mais útil para modelos de regressão linear, pois pode ajudá-lo a encontrar estimativas de valor real imparciais para seu alfa e beta.

```python
import plotly.graph_objects as go
import statsmodels.api as sm
import pandas as pd

# Suponha que você tenha seus dados no DataFrame df
# Substitua as variáveis abaixo pelos nomes das suas colunas
XI = df_1[['Indústria - total']]  # Transforma a série em DataFrame
yI = df_1['AUTOM_IND']

# Adicione uma constante à variável independente
X = sm.add_constant(XI)

# Crie o modelo
modelo = sm.OLS(yI, X)

# Treine o modelo
resultado = modelo.fit()

# Exiba o resumo do modelo
print(resultado.summary())

# Crie um gráfico de dispersão
fig_mínimosquad = go.Figure()

fig_mínimosquad.add_trace(go.Scatter(x=XI['Indústria - total'], y=yI, mode='markers', name='Dados'))

# Adicione a linha de regressão
fig_mínimosquad.add_trace(go.Scatter(x=XI['Indústria - total'], y=resultado.predict(X),
                         mode='lines', name='Reta de Regressão'))

# Adicione o resumo do modelo como anotação no gráfico
resumo_texto = str(resultado.summary())
fig_mínimosquad.add_annotation(text=resumo_texto, xref="paper", yref="paper", x=0.5, y=0.8,
                   showarrow=False, font=dict(size=10), bgcolor="lightyellow")

# Atualize o layout do gráfico
fig_mínimosquad.update_layout(title='Regressão Linear - Resumo do Modelo Mínimos Quadrados',
                  xaxis_title='PIB DA INDÚSTRIA (MILHÕES DE REAIS)',
                  yaxis_title='AUTOMAÇÃO INDUSTRIAL (MILHÕES DE REAIS)')

# Exiba o gráfic
fig_mínimosquad.show()

```

```python
# Calcular o viés da previsão
viés = data3['Previsão'] - data3['AUTOM_IND']
viés_médio = viés.mean()

# Calcular a razão do viés
média_observado = data3['AUTOM_IND'].mean()
razão_viés = viés_médio / média_observado

print(viés_médio)
print(razão_viés)
```

**Viés médio** é uma medida da diferença entre a média da distribuição da estimativa e o valor verdadeiro do parâmetro. Em outras palavras, é uma medida de quanto a estimativa está enviesada.

**Razão do viés** é a razão entre o viés médio e o valor verdadeiro do parâmetro. É uma medida da magnitude do viés.

No seu caso, o **viés médio** é de -0,14814814814814814. Isso significa que, em média, a estimativa está 0,14814814814814814 abaixo do valor verdadeiro do parâmetro.

A **razão do viés** é de -4,476876930653176e-05. Isso significa que o viés é muito pequeno em relação ao valor verdadeiro do parâmetro.

Em outras palavras, a estimativa está um pouco abaixo do valor verdadeiro do parâmetro, mas a diferença é muito pequena.

Aqui estão alguns exemplos para ilustrar o significado do viés médio e da razão do viés:

* **Viés médio = 0** significa que a estimativa é não enviesada.
* **Viés médio > 0** significa que a estimativa está tendenciosa para cima.
* **Viés médio < 0** significa que a estimativa está tendenciosa para baixo.

* **Razão do viés = 0** significa que o viés é zero.
* **Razão do viés > 0** significa que o viés é significativo.
* **Razão do viés < 0** significa que o viés é insignificante.

No seu caso, o fato de o viés médio ser pequeno e a razão do viés ser insignificante sugere que a estimativa é relativamente precisa.
