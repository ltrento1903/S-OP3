```python
import dash
from dash import dcc, html
import dash_bootstrap_components as dbc

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# Dados do carrossel
carousel_items = [
    {
        "key": "1",
        "src": "/static/images/slide1.svg",
        "header": "Com cabeçalho",
        "caption": "e legenda",
    },
    {
        "key": "2",
        "src": "/static/images/slide2.svg",
        "header": "Apenas com cabeçalho",
        "caption": "",
    },
    {
        "key": "3",
        "src": "/static/images/slide3.svg",
        "header": "",
        "caption": "Este slide possui apenas uma legenda",
    },
]

# Layout do aplicativo
app.layout = html.Div(
    [
        dbc.Carousel(
            items=[
                html.Div(
                    [
                        html.Img(src=item["src"]),
                        html.H3(item["header"]),
                        html.P(item["caption"]),
                    ]
                )
                for item in carousel_items
            ],
            id="carousel",
        ),
        html.Div(id="carousel-output"),
    ]
)

if __name__ == "__main__":
    app.run_server(debug=True)
```
